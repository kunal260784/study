```python
from langchain.tools import tool
from langchain.chat_models import init_chat_model
from langchain.messages import SystemMessage, AIMessage, HumanMessage
from pprint import pprint
```

```python
@tool
def add(a:float, b:float) -> float:

    """
    add Function is created to perform addition of two Numbers.
    
    :param a: First Number to Perform Addition
    :type a: float
    :param b: Second Number to Perform Addition
    :type b: float

    :return: Sum of given two Numbers
    :rtype: float

    ..code-block :: python
        #example

        >>> add(1.0,2.3)
        3.3

        >>> add(2,5)
        7.0
    """
    return a + b



@tool
def Subtract(a:float, b:float) -> float:

    """
    Subtract Function is created to perform Subtraction of two Numbers.
    
    :param a: First Number to Perform Addition
    :type a: float
    :param b: Second Number to Perform Addition
    :type b: float

    :return: Subtract second number from first one.
    :rtype: float

    ..code-block :: python
        #example

        >>> Subtract(1.0,2.3)
        -1.3

        >>> Subtract(5,1)
        4.0
    """
    return a - b
```

```python
llm = init_chat_model(
    model="gpt-4o-mini",
    model_provider="openai"
)

llm_tool = llm.bind_tools([add,Subtract], tool_choice="any")
```

```python
result = llm_tool.invoke("multiply 2 and 3")
result.pretty_print()
```

```python
result = llm.invoke("Subtract 2 from 1.5")
print(result)
```
