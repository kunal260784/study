```python
from langchain.tools import tool
from langchain.chat_models import init_chat_model
from dotenv import load_dotenv
load_dotenv()
import os
```

```python
os.environ['OPENAI_API_KEY'] = os.getenv('OPENAI_API_KEY')
```

```python
@tool
def add(a:float, b:float ) -> float:

    """
    Function is used to add two number
    
    :param a: First Parameter to perform addition
    :type a: float
    :param b: Second Parameter to perform addition
    :type b: float

    :return: Returns the Sum of First and Second Parameter
    :rtype: float

        ..code-block :: python
            #example

            >>> add(1.0,2.3)
            3.3

            >>> add(2,5)
            7.0
    """

    return a + b
```

```python
print(add.name)
```

```python
llm = init_chat_model(
    model="gpt-4.1-mini",
    model_provider="openai"
)
llm_tool = llm.bind_tools([add], tool_choice="any")
```

```python
llm_tool.invoke("add 5 and 1.1")
```
