```python
from langchain.chat_models import init_chat_model
from langchain_community.tools import WikipediaQueryRun
from langchain_tavily import TavilySearch
from langchain_community.utilities import WikipediaAPIWrapper
from langchain.agents import create_agent
from dotenv import load_dotenv
import os
from pprint import pprint
```

```python
os.environ['OPENAI_API_KEY'] = os.getenv('OPENAI_API_KEY')
```

```python
wikitool = WikipediaQueryRun(api_wrapper=WikipediaAPIWrapper(
    top_k_results=5,
    doc_content_chars_max=2000
))

tavilytool = TavilySearch(
    max_results = 5,
    topic = "general"
)
```

```python
llm = init_chat_model(model="gpt-4o-mini", model_provider="openai")
```

```python
llmmodel = create_agent(model=llm, tools=[wikitool, tavilytool],
                        system_prompt="You are a Power BI Expert and can only answer Power BI related to Questions")
```

```python
result = llmmodel.invoke({"messages":"What is llm"})
```

```python
for item in result['messages']:
    item.pretty_print()
```
